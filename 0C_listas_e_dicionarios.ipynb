## Listas: o conceito

Uma **lista** é uma coleção ordenada de elementos. Você cria uma lista usando colchetes `[]` e separando os elementos com vírgula:

````python
municipios = ["São Mateus", "Linhares", "Jaguaré", "Nova Venécia", "Sooretama"]
sacas_por_mes = [120, 85, 145, 210, 175, 90]

print(municipios)
print(sacas_por_mes)
````

**Saída esperada:**

```
['São Mateus', 'Linhares', 'Jaguaré', 'Nova Venécia', 'Sooretama']
[120, 85, 145, 210, 175, 90]
```

Listas podem conter qualquer tipo de dado: strings, números, booleanos, e até misturar tipos. Mas, na prática, costumamos manter listas com um único tipo porque facilita o processamento depois.

## Acessando elementos: indexação

Cada elemento de uma lista tem uma **posição**, chamada de **índice**. Em Python, os índices começam em **zero**: o primeiro elemento está no índice `0`, o segundo no `1`, e assim por diante.

Você pega um elemento usando colchetes com o índice:

````python
municipios = ["São Mateus", "Linhares", "Jaguaré", "Nova Venécia", "Sooretama"]

print(municipios[0])
print(municipios[2])
print(municipios[-1])
print(municipios[-2])
````

**Saída esperada:**

```
São Mateus
Jaguaré
Sooretama
Nova Venécia
```

**Dica:** índices negativos contam de trás pra frente. `-1` é sempre o último elemento, `-2` o penúltimo, e por aí vai. Muito útil quando você não sabe o tamanho da lista mas quer o último item.

## Fatiamento (slicing)

Muitas vezes você não quer um elemento só, quer um **pedaço** da lista. Pra isso, usa a sintaxe `lista[inicio:fim]`, onde `inicio` é o índice onde começa (incluído) e `fim` é o índice onde para (**não** incluído).

````python
sacas_por_mes = [120, 85, 145, 210, 175, 90, 130, 200, 180, 95, 110, 160]

primeiro_trimestre = sacas_por_mes[0:3]
segundo_trimestre = sacas_por_mes[3:6]
segundo_semestre = sacas_por_mes[6:]
ultimos_tres = sacas_por_mes[-3:]

print(primeiro_trimestre)
print(segundo_trimestre)
print(segundo_semestre)
print(ultimos_tres)
````

**Saída esperada:**

```
[120, 85, 145]
[210, 175, 90]
[130, 200, 180, 95, 110, 160]
[95, 110, 160]
```

**Atenção:** o `fim` é **exclusivo**. `sacas_por_mes[0:3]` devolve os elementos dos índices 0, 1 e 2, ou seja, três elementos, mas não inclui o índice 3. Essa é provavelmente a pegadinha mais comum de fatiamento em Python.

Se você omitir o início, o fatiamento começa do primeiro elemento. Se omitir o fim, vai até o último.

## Métodos essenciais: append, sort e len

Listas são **mutáveis**, ou seja, você pode mexer nelas depois de criadas. Os três recursos mais usados no dia a dia são:

| Recurso | O que faz |
|---------|-----------|
| `lista.append(x)` | Adiciona `x` no final da lista |
| `lista.sort()` | Ordena a lista em ordem crescente (altera a própria lista) |
| `len(lista)` | Devolve a quantidade de elementos (é uma função, não um método) |

Na prática:

````python
municipios = ["São Mateus", "Linhares", "Jaguaré"]

print(len(municipios))

municipios.append("Nova Venécia")
print(municipios)

municipios.sort()
print(municipios)

print(len(municipios))
````

**Saída esperada:**

```
3
['São Mateus', 'Linhares', 'Jaguaré', 'Nova Venécia']
['Jaguaré', 'Linhares', 'Nova Venécia', 'São Mateus']
4
```

**Dica:** note que `sort()` **não devolve nada**, ele ordena a lista no lugar. Por isso a gente escreve `municipios.sort()` numa linha e depois `print(municipios)` na próxima, e não `print(municipios.sort())` (que imprimiria `None`).

## Dicionários: o conceito

Uma **lista** guarda elementos acessados por posição. Um **dicionário** guarda pares de **chave e valor**: você acessa os valores pela chave, não por um número.

Dicionários são perfeitos pra representar um "registro" com vários campos, tipo uma linha de um cadastro, os dados de um município, ou as propriedades de uma fazenda.

Você cria um dicionário usando chaves `{}` e separando os pares com vírgula, no formato `chave: valor`:

````python
sao_mateus = {
    "nome": "São Mateus",
    "uf": "ES",
    "sacas_2024": 45000,
    "hectares": 1400,
    "variedade": "conilon",
    "eh_litoraneo": True
}

print(sao_mateus)
````

**Saída esperada:**

```
{'nome': 'São Mateus', 'uf': 'ES', 'sacas_2024': 45000, 'hectares': 1400, 'variedade': 'conilon', 'eh_litoraneo': True}
```

As chaves normalmente são strings, e os valores podem ser de qualquer tipo: números, strings, booleanos e até outras listas ou dicionários.

## Acessando valores: colchetes e .get()

Pra pegar um valor do dicionário, você usa colchetes com a chave (parecido com lista, só que passando a chave no lugar do índice):

````python
sao_mateus = {
    "nome": "São Mateus",
    "uf": "ES",
    "sacas_2024": 45000,
    "hectares": 1400
}

print(sao_mateus["nome"])
print(sao_mateus["sacas_2024"])
````

**Saída esperada:**

```
São Mateus
45000
```

Tem um perigo nessa forma: se você tentar acessar uma chave que **não existe**, o Python dá erro (`KeyError`) e para a execução. Pra acessar com segurança, use o método `.get(chave)`:

````python
sao_mateus = {
    "nome": "São Mateus",
    "sacas_2024": 45000
}

print(sao_mateus.get("nome"))
print(sao_mateus.get("populacao"))
print(sao_mateus.get("populacao", "não informado"))
````

**Saída esperada:**

```
São Mateus
None
não informado
```

**Dica:** `.get()` devolve `None` quando a chave não existe, ou um valor padrão se você passar um segundo argumento. É a forma segura de mexer com dicionários quando você não sabe se todas as chaves estão lá, o que é muito comum quando os dados vêm de fora.

## Iterando sobre dicionários: .items()

Quase sempre a gente quer percorrer todo o dicionário, pegando cada par chave-valor. O método `.items()` devolve esses pares prontos pra usar num loop `for`:

````python
sao_mateus = {
    "nome": "São Mateus",
    "uf": "ES",
    "sacas_2024": 45000,
    "hectares": 1400,
    "variedade": "conilon"
}

for chave, valor in sao_mateus.items():
    print(f"{chave}: {valor}")
````

**Saída esperada:**

```
nome: São Mateus
uf: ES
sacas_2024: 45000
hectares: 1400
variedade: conilon
```

Esse padrão (`for chave, valor in dicionario.items()`) é um dos mais usados em Python. Grava ele.
