## O que é o Google Colab?

O Google Colaboratory, ou simplesmente **Colab**, é um ambiente gratuito oferecido pelo Google que permite escrever e executar código Python diretamente no navegador, sem precisar instalar nada no seu computador.

Ele é baseado nos famosos **Jupyter Notebooks** e é amplamente utilizado em cursos e projetos de Ciência de Dados e Inteligência Artificial no mundo todo.

**Por que vamos usar o Colab neste curso?**

- Não precisa instalar Python nem nenhuma ferramenta no seu computador
- Já vem com as principais bibliotecas que vamos usar (Pandas, Matplotlib, Seaborn, NumPy, Scikit-learn)
- Funciona em qualquer computador com navegador e internet
- Salva automaticamente no seu Google Drive
- É gratuito

## Acessando o Google Colab

Para acessar o Colab, basta entrar no endereço:

👉 **https://colab.research.google.com**

Você precisará de uma conta Google (Gmail). Ao acessar, verá a tela inicial com opções para criar um novo notebook ou abrir um existente.

## Criando um novo notebook

Na tela inicial do Colab, clique em **"Novo notebook"** (ou *"New notebook"*, se estiver em inglês). Um notebook em branco será criado, pronto para uso.

O notebook recém-criado recebe um nome padrão como `Untitled0.ipynb`. Para renomear, basta clicar sobre o nome no canto superior esquerdo e digitar o novo nome.

**Dica:** use nomes descritivos para seus notebooks. Por exemplo: `0A_Google_Colab.ipynb`

## A interface do Colab

A interface do Colab tem as seguintes áreas principais:

- **Barra de menus** (topo): Arquivo, Editar, Ver, Inserir, Ambiente de execução, Ferramentas, Ajuda
- **Barra de ferramentas**: botões rápidos para ações comuns
- **Painel lateral esquerdo**: acesso a arquivos, busca, variáveis e snippets de código
- **Área central**: onde ficam as células do notebook (é aqui que a mágica acontece!)

## Células: o conceito fundamental

Um notebook é formado por **células**. Existem dois tipos:

### Células de código

São células onde você escreve e executa código Python. Elas têm um fundo levemente acinzentado e um botão de play (▶) no lado esquerdo.

Exemplo de célula de código:

````python
# Isso é uma célula de código
# Tudo que você escrever aqui será executado como Python
print("Olá, Colab!")
````

**Saída esperada:**
```
Olá, Colab!
```

### Células de texto (Markdown)

São células onde você escreve texto formatado, explicações, títulos e anotações. Elas usam a linguagem **Markdown** para formatação.

Exemplo de célula de texto:

```markdown
## Minha primeira análise
Este notebook contém a análise exploratória do dataset de **café do Espírito Santo**.
```

Para alternar entre os tipos de célula, use o menu **Inserir → Célula de código** ou **Inserir → Célula de texto**.

## Executando células de código

Existem três formas principais de executar uma célula de código:

| Ação | Como fazer |
|------|-----------|
| Executar a célula atual | Clique no botão ▶ à esquerda da célula |
| Executar e ir para a próxima | `Shift + Enter` |
| Executar e permanecer na célula | `Ctrl + Enter` |

Vamos praticar! Execute a célula abaixo:

````python
2 + 3
````

**Saída esperada:**
```
5
```

Agora tente esta:

````python
nome = "Análise Exploratória de Dados"
print(f"Bem-vindo à disciplina de {nome}!")
````

**Saída esperada:**
```
Bem-vindo à disciplina de Análise Exploratória de Dados!
```

**Importante:** as células são executadas na ordem em que você roda, não na ordem em que aparecem no notebook. Se você executar a célula 3 antes da célula 2, e a célula 3 depender de algo definido na célula 2, vai dar erro. Na dúvida, use o menu **Ambiente de execução → Executar tudo** para rodar todas em sequência.

## Upload de arquivos

Em Análise de Dados, frequentemente precisamos carregar arquivos (CSV, Excel, etc.) para dentro do Colab. Existem duas formas principais:

### Forma 1: Upload temporário

Clique no ícone de **pasta (📁)** no painel lateral esquerdo e depois no ícone de **upload (⬆)**. Selecione o arquivo do seu computador.

Você também pode fazer upload por código:

```python
from google.colab import files

arquivo = files.upload()
```

Ao executar, aparecerá um botão para selecionar o arquivo. Depois de carregado, ele ficará disponível na pasta `/content/`.

**Atenção:** arquivos enviados por upload temporário são perdidos quando o notebook é desconectado. Para arquivos permanentes, use o Google Drive.

### Forma 2: Conectando ao Google Drive (recomendado)

Esta é a forma mais prática para trabalhar com arquivos de forma permanente:

```python
from google.colab import drive

drive.mount('/content/drive')
```

Ao executar, o Colab pedirá permissão para acessar seu Drive. Após autorizar, seus arquivos estarão acessíveis em `/content/drive/MyDrive/`.

Exemplo - lendo um arquivo CSV que está no seu Drive (trecho **ilustrativo**, requer um arquivo real no caminho indicado):

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/AED_IA/dataset.csv')
df.head()
```

**Dica:** crie uma pasta no seu Drive chamada `AED_IA` para organizar os arquivos da disciplina.
