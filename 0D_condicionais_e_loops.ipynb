## Condicionais: if, elif, else

Uma **condicional** permite que seu código escolha um caminho com base numa comparação. A estrutura básica é:

```
if condicao:
    ...código que roda se a condição for verdadeira...
elif outra_condicao:
    ...código que roda se a primeira foi falsa e essa for verdadeira...
else:
    ...código que roda se todas as anteriores foram falsas...
```

Na prática, vamos classificar a produtividade de uma fazenda de café:

````python
produtividade = 32.5

if produtividade >= 40:
    classificacao = "alta"
elif produtividade >= 25:
    classificacao = "média"
else:
    classificacao = "baixa"

print(f"Produtividade de {produtividade} sacas/hectare: {classificacao}")
````

**Saída esperada:**

```
Produtividade de 32.5 sacas/hectare: média
```

**Regras importantes:**

- Use `:` no fim da linha da condição
- O código dentro do bloco precisa estar **indentado** (4 espaços é o padrão)
- `elif` e `else` são opcionais: você pode ter só `if`, ou `if` com vários `elif`, ou `if` com `else` direto

**Atenção:** indentação no Python não é enfeite, é **sintaxe**. Se você indentar errado, o código não roda, ou roda diferente do que você esperava.

## Operadores lógicos: and, or, not

Às vezes uma condição só não basta: você quer combinar várias. Pra isso existem três operadores lógicos:

| Operador | O que faz | Exemplo |
|----------|-----------|---------|
| `and` | Verdadeiro só se **as duas** forem verdadeiras | `nota >= 7 and frequencia >= 0.75` |
| `or`  | Verdadeiro se **pelo menos uma** for verdadeira | `variedade == "conilon" or variedade == "arabica"` |
| `not` | Inverte o valor lógico (vira o oposto) | `not em_safra` |

Vamos combinar três condições pra decidir se uma fazenda se qualifica pra um programa de incentivo:

````python
hectares = 120
produtividade = 35
variedade = "conilon"

qualifica = hectares >= 50 and produtividade >= 30 and variedade == "conilon"

print(f"Hectares: {hectares}")
print(f"Produtividade: {produtividade}")
print(f"Variedade: {variedade}")
print(f"Qualifica para o programa? {qualifica}")
````

**Saída esperada:**

```
Hectares: 120
Produtividade: 35
Variedade: conilon
Qualifica para o programa? True
```

**Dica:** esses operadores retornam `True` ou `False`, ou seja, um `bool`. Por isso você pode guardar o resultado numa variável (como fizemos com `qualifica`) e depois usar essa variável direto num `if`.

## Loop for sobre listas

Quando você tem uma lista e quer fazer alguma coisa **com cada elemento**, o jeito limpo é usar um loop `for`. A estrutura é:

```
for variavel in lista:
    ...código que roda uma vez pra cada elemento...
```

O `for` pega cada elemento da lista, um por um, e guarda em `variavel` pra você usar dentro do bloco.

````python
municipios = ["São Mateus", "Linhares", "Jaguaré", "Nova Venécia"]

for municipio in municipios:
    print(f"Processando {municipio}...")
````

**Saída esperada:**

```
Processando São Mateus...
Processando Linhares...
Processando Jaguaré...
Processando Nova Venécia...
```

Outro padrão muito comum é percorrer uma lista de números e ir acumulando um total. Por exemplo, somar as sacas produzidas ao longo dos meses de uma safra:

````python
sacas_por_mes = [120, 85, 145, 210, 175, 90]
total = 0

for sacas in sacas_por_mes:
    total = total + sacas

print(f"Total da safra: {total} sacas")
````

**Saída esperada:**

```
Total da safra: 825 sacas
```

**Dica:** esse padrão de "inicializar uma variável em 0 fora do loop e ir somando dentro" é clássico. No Pandas a gente raramente faz isso na mão (existe `sum()`, `mean()`, etc.), mas é importante entender o que está acontecendo por baixo.

## A função range() e for com range

Nem sempre você quer iterar sobre uma lista existente. Às vezes você só quer "fazer X vezes" ou "contar de 1 até N". Pra isso existe a função `range()`, que gera uma sequência de números:

| Chamada | Gera |
|---------|------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |

O último número é **sempre exclusivo** (não é incluído).

Combinando `for` com `range()`, você pode repetir uma ação um número exato de vezes. Exemplo: imprimir as safras dos últimos 5 anos, começando em 2020:

````python
for ano in range(2020, 2025):
    print(f"Safra de {ano}")
````

**Saída esperada:**

```
Safra de 2020
Safra de 2021
Safra de 2022
Safra de 2023
Safra de 2024
```

## Loop while

O `for` é ótimo quando você sabe sobre o que vai iterar (uma lista, um range). Mas às vezes você não sabe quantas repetições vai precisar, você só sabe que quer repetir **enquanto** alguma condição for verdadeira. Pra isso existe o `while`:

```
while condicao:
    ...código que roda enquanto a condição for verdadeira...
```

Exemplo: acumular sacas de várias colheitas até bater uma meta de 500 sacas:

````python
meta = 500
acumulado = 0
colheita = 0

while acumulado < meta:
    colheita = colheita + 1
    acumulado = acumulado + 120
    print(f"Colheita {colheita}: acumulado = {acumulado} sacas")

print(f"Meta batida em {colheita} colheitas")
````

**Saída esperada:**

```
Colheita 1: acumulado = 120 sacas
Colheita 2: acumulado = 240 sacas
Colheita 3: acumulado = 360 sacas
Colheita 4: acumulado = 480 sacas
Colheita 5: acumulado = 600 sacas
Meta batida em 5 colheitas
```

**Atenção:** se a condição do `while` nunca se tornar falsa, o loop roda pra sempre (**loop infinito**). Garanta que algo dentro do loop atualize a variável da condição. No exemplo acima, a linha `acumulado = acumulado + 120` é o que eventualmente faz `acumulado < meta` virar falso e o loop parar.

## Operador in

O operador `in` verifica se um elemento está dentro de uma coleção (lista, string, dicionário). O resultado é `True` ou `False`:

````python
municipios = ["São Mateus", "Linhares", "Jaguaré", "Nova Venécia"]

print("São Mateus" in municipios)
print("Vitória" in municipios)

fazenda = {
    "nome": "Fazenda Santa Clara",
    "hectares": 80,
    "variedade": "conilon"
}

print("nome" in fazenda)
print("produtor" in fazenda)
````

**Saída esperada:**

```
True
False
True
False
```

**Dica:** em dicionários, `in` checa as **chaves**, não os valores. É o parceiro natural do `.get()` que você viu no notebook anterior: antes de acessar uma chave, você pode garantir que ela existe com `if chave in dicionario:`. Esse é o padrão mais comum pra "só faz alguma coisa se o campo estiver preenchido".

O `in` também é a mesma palavra-chave que aparece no `for`: `for municipio in municipios` usa essa mesma ideia de "cada elemento dentro da coleção".
